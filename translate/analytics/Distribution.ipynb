```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def draw_histogram(filename, bin_range, bin_width, dataset_name):
    """
    绘制频率分布直方图

    Args:
        filename: 数据文件路径
        bin_range: 区间范围，一个长度为2的列表，表示最小值和最大值
        bin_width: 区间间隔

    Returns:
        None
    """

    # 读取数据
    data = np.loadtxt(filename)

    # 计算bin的个数
    num_bins = int((bin_range[1] - bin_range[0]) / bin_width)

    # 绘制直方图
    plt.hist(data, bins=num_bins, range=bin_range)

    # 设置标题和坐标轴标签
    plt.title('Quality Distribution of ' + dataset_name)
    plt.xlabel('Quality Score (cos-sim of embeddings for each pair)')
    plt.ylabel('Frequency')

    # 显示图形
    plt.show()
```

```python
# 示例用法
filename = './ccmatrix/CCMatrix.en-zh.sample.sim'  # 替换为你的文件名
bin_range = [0.2, 1]  # 设置区间范围
bin_width = 0.025  # 设置区间间隔

draw_histogram(filename, bin_range, bin_width, "CCMatrix (ZH-EN)")
```

```python
# 示例用法
filename = './translation2019/result.sim'  # 替换为你的文件名
bin_range = [0.2, 1]  # 设置区间范围
bin_width = 0.025  # 设置区间间隔

draw_histogram(filename, bin_range, bin_width, "nlp_chinese_corpus/translation2019")
```
